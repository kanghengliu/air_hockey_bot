```python
import pandas as pd

# Function to load the data and calculate the average total reward for each group of 100,000 rows
def calculate_group_averages(file_path, group_size=100000, column_name='Total Reward'):
    # Load the data
    df = pd.read_csv(file_path)

    # Calculate the number of groups
    num_groups = len(df) // group_size + (1 if len(df) % group_size else 0)

    # Calculate the average total reward for each group
    group_averages = pd.DataFrame({
        'Group': [f"{i * group_size}-{(i + 1) * group_size - 1}" for i in range(num_groups)],
        'Average Total Reward': [
            df[column_name][i * group_size:(i + 1) * group_size].mean()
            for i in range(num_groups)
        ]
    })

    return group_averages

# Assuming the CSV file is named 'training_log.csv' and is in the same directory as the script
group_averages = calculate_group_averages('checkpoints/training_log.csv')
group_averages
```
